```python
def train_test_split(X, y, test_size=0.2, shuffle=True):
    assert len(X) == len(y)
    indices = np.arange(len(X))
    if shuffle:
        np.random.shuffle(indices)
    split_idx = int(len(X) * (1 - test_size))
    train_idx, test_idx = indices[:split_idx], indices[split_idx:]
    # Use iloc to index by row numbers
    return X.iloc[train_idx], X.iloc[test_idx], y.iloc[train_idx], y.iloc[test_idx]

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
def min_max_scale_fit(X):
    """Fit function: calculate min and max for each column"""
    min_vals = np.min(X, axis=0)
    max_vals = np.max(X, axis=0)
    return min_vals, max_vals

def min_max_scale_transform(X, min_vals, max_vals):
    """Transform function: scale using the previously calculated min and max"""
    denom = max_vals - min_vals
    denom[denom == 0] = 1  # Prevent division by zero
    return (X - min_vals) / denom

# Convert to NumPy arrays
X_train_np = X_train.to_numpy()
X_test_np = X_test.to_numpy()

# Fit on training data
min_vals, max_vals = min_max_scale_fit(X_train_np)

# Transform training and testing data using same scaling
X_train_scaled = min_max_scale_transform(X_train_np, min_vals, max_vals)
X_test_scaled = min_max_scale_transform(X_test_np, min_vals, max_vals)
```

```python
import numpy as np

def pca(X, n_components=2, fit=True):
    """
    Performs Principal Component Analysis (PCA) on the data.

    Args:
        X (numpy.ndarray): The input data.
        n_components (int, optional): Number of principal components to keep. Defaults to 2.
        fit (bool, optional): Whether to fit the PCA model. Defaults to True.

    Returns:
        numpy.ndarray: The transformed data.
        numpy.ndarray: The principal components (eigenvectors).
    """

    X_meaned = X - np.mean(X, axis=0)
    cov_mat = np.cov(X_meaned, rowvar=False)
    eig_vals, eig_vecs = np.linalg.eigh(cov_mat)
    sorted_indices = np.argsort(eig_vals)[::-1]
    principal_components = eig_vecs[:, sorted_indices]
    principal_components = principal_components[:, :n_components]

    #Removed self here as it was not needed since the function is not a method within a class
    #if fit:  # Fit the model by storing principal components
    #    self.principal_components = principal_components

    transformed_data = np.dot(X_meaned, principal_components)  # Transform the data

    if fit:
        return transformed_data, principal_components
    else:
        return transformed_data


# Usage:
# Fit and transform the training data
X_train_pca, principal_components = pca(X_train_scaled, n_components=2, fit=True)

# Transform the test data using the fitted principal components
X_test_pca = pca(X_test_scaled, n_components=2, fit=False)
```

```python
class GaussianNaiveBayes:
    def __init__(self):
        self.classes = None
        self.mean = None
        self.variance = None
        self.prior = None

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.mean = np.zeros((len(self.classes), X.shape[1]))
        self.variance = np.zeros((len(self.classes), X.shape[1]))
        self.prior = np.zeros(len(self.classes))

        for i, c in enumerate(self.classes):
            X_c = X[y == c]
            self.mean[i, :] = X_c.mean(axis=0)
            self.variance[i, :] = X_c.var(axis=0) + 1e-6
            self.prior[i] = X_c.shape[0] / X.shape[0]

    def predict(self, X):
        return np.array([self._predict_single(x) for x in X])

    def _predict_single(self, x):
        posteriors = []
        for i, c in enumerate(self.classes):
            prior = np.log(self.prior[i])
            likelihood = np.sum(np.log(self._gaussian_pdf(x, self.mean[i], self.variance[i])))
            posteriors.append(prior + likelihood)
        return self.classes[np.argmax(posteriors)]

    def _gaussian_pdf(self, x, mean, var):
     eps = 1e-9  # Avoid division by 0 and log(0)
     coef = 1.0 / np.sqrt(2.0 * np.pi * var)
     exponent = np.exp(-((x - mean) ** 2) / (2 * var))
     return np.clip(coef * exponent, eps, None)
```

```python
# Node class for the Decision Tree
class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value

# Decision Tree Class (Custom C4.5 style, use gini impurity of cart style)
class DecisionTree:
    def __init__(self, max_depth=None, min_samples_split=2, **kwargs):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.root = None

    def _gini_impurity(self, y):
        unique_classes, counts = np.unique(y, return_counts=True)
        impurity = 1 - np.sum((counts / len(y)) ** 2)
        return impurity

    def _best_split(self, X, y):
        best_feature, best_threshold = None, None
        best_gini = float('inf')

        # Get the number of features (columns) in X
        num_features = X.shape[1]

        for feature in range(num_features):
            # Get unique values for the current feature
            # Use X[:, feature] to access the feature column for NumPy arrays
            thresholds = np.unique(X[:, feature])

            for threshold in thresholds:
                # Use boolean indexing for NumPy arrays
                left_idx = X[:, feature] <= threshold
                right_idx = X[:, feature] > threshold

                if len(y[left_idx]) >= self.min_samples_split and len(y[right_idx]) >= self.min_samples_split:
                    gini_left = self._gini_impurity(y[left_idx])
                    gini_right = self._gini_impurity(y[right_idx])
                    gini_split = (len(y[left_idx]) / len(y)) * gini_left + (len(y[right_idx]) / len(y)) * gini_right
                    if gini_split < best_gini:
                        best_gini = gini_split
                        best_feature = feature
                        best_threshold = threshold
        return best_feature, best_threshold

    def _build_tree(self, X, y, depth=0):
        if depth == self.max_depth or len(np.unique(y)) == 1 or len(y) < self.min_samples_split:
            most_common = np.bincount(y).argmax()
            return Node(value=most_common)

        best_feature, best_threshold = self._best_split(X, y)
        if best_feature is None:
            most_common = np.bincount(y).argmax()
            return Node(value=most_common)

        # Use boolean indexing for NumPy arrays
        left_idx = X[:, best_feature] <= best_threshold
        right_idx = X[:, best_feature] > best_threshold

        left_subtree = self._build_tree(X[left_idx], y[left_idx], depth + 1)
        right_subtree = self._build_tree(X[right_idx], y[right_idx], depth + 1)

        return Node(feature=best_feature, threshold=best_threshold, left=left_subtree, right=right_subtree)

    def fit(self, X, y):
        self.root = self._build_tree(X, y)

    def _predict_single(self, x, node):
        if node.value is not None:
            return node.value
        # Access feature using NumPy array indexing
        if x[node.feature] <= node.threshold:
            return self._predict_single(x, node.left)
        else:
            return self._predict_single(x, node.right)
    def predict(self, X):
        return np.array([self._predict_single(x, self.root) for x in X]) # Changed to iterate over rows of X
```

```python
def accuracy_score(y_true, y_pred):
    return (y_true == y_pred).sum() / len(y_true)

def precision_score(y_true, y_pred):
    TP = ((y_true == 1) & (y_pred == 1)).sum()
    FP = ((y_true == 0) & (y_pred == 1)).sum()
    # Return 1 if denominator is 0 to avoid division by zero error
    return TP / (TP + FP) if (TP + FP) != 0 else 1

def recall_score(y_true, y_pred):
    TP = ((y_true == 1) & (y_pred == 1)).sum()
    FN = ((y_true == 1) & (y_pred == 0)).sum()
    # Return 1 if denominator is 0 to avoid division by zero error
    return TP / (TP + FN) if (TP + FN) != 0 else 1

def f1_score(y_true, y_pred):
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    # Return 1 if denominator is 0 to avoid division by zero error
    return 2 * prec * rec / (prec + rec) if (prec + rec) != 0 else 1
```

```python
def plot_confusion_matrix(y_true, y_pred, labels):
    import numpy as np
    import matplotlib.pyplot as plt

    matrix = np.zeros((2, 2), dtype=int)
    for t, p in zip(y_true, y_pred):
        matrix[int(t)][int(p)] += 1

    fig, ax = plt.subplots()
    cax = ax.matshow(matrix, cmap='Blues')
    plt.colorbar(cax)
    for (i, j), val in np.ndenumerate(matrix):
        ax.text(j, i, f'{val}', ha='center', va='center')

    ax.set_xticklabels([''] + labels)
    ax.set_yticklabels([''] + labels)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()

    # Return the confusion matrix
    return matrix
```
